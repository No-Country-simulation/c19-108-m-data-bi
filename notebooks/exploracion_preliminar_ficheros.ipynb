```python
# encontramos que existe un problema de encoding en algunos ficheros
# para explorarlos a fines practicos para el analisis preliminar, trataremos las excepciones buscando el encoding que se adecúe
```

```python
for df in lista_produccion:
    with open(f'../raw/{df}', 'rb') as f: # abrimos el fuchero y lo leemos como dato binario
      result = chardet.detect(f.read()) # identificamos el tipo de encoding en que estan los datos
      encoding = result['encoding'] # obtenemos el valor de encoding que nos interesa y lo instanciamos a una variable
    data = pd.read_csv(f'../raw/{df}', encoding= encoding)
    informe = r.informe_dataframe(data)
    print(informe)
```
